```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pot1(x):
    G = 1.0
    M = 1.0
    B = 1.0
    return - ((G * M) / B) * np.arcsin(1 / np.cosh(x))

def pot2(x):
    G = 1.0
    M = 1.0
    return - ((G * M) / x)


def func1(x):
    return np.arcsin(1 / np.cosh(x))

def func2(x):
    return 1 / np.cosh(x)

def func3(x):
    return np.arcsin(x)

x_small = np.linspace(0.1, 2 * np.pi, num=100)
x2_small = np.linspace(0, 1, num=100)
x_large = np.linspace(0.1, 100, num=1000)
x2_large = np.linspace(0, 1, num=1000)
```

```python
plt.plot(x_small, pot1(x_small), label='$\Phi(x)$ Homeoid')
plt.plot(x_small, pot2(x_small), label='$\Phi(x)$ Point Mass')
plt.plot(x_small, func1(x_small), label='arcsin(sech(x))')
plt.plot(x_small, func2(x_small), label='sech(x)')
plt.plot(x2_small, func3(x2_small), label='arcsin(x)')
plt.title('Function behavior small radius')
plt.ylim([-4, 2])
plt.xlabel('Radius')
plt.ylabel('Output')
plt.legend()
plt.savefig('ay630_hw2_problem1_smallr')
plt.show()
```

```python
plt.plot(x_large, pot1(x_large), label='$\Phi(x)$ Homeoid')
plt.plot(x_large, pot2(x_large), label='$\Phi(x)$ Point Mass')
#plt.plot(x_large, func1(x_large), label='arcsin(sech(x))')
#plt.plot(x_large, func2(x_large), label='sech(x)')
#plt.plot(x2_large, func3(x2_large), label='arcsin(x)')
plt.title('Function behavior larger radius')
plt.ylim([-4, 2])
plt.xlabel('Radius')
plt.ylabel('Output')
plt.legend()
plt.savefig('ay630_hw2_problem1_larger')
plt.show()
```
